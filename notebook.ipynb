# BOOKS PUBLISHED IN YEAR

# Fiction

```python
fic = booklist[booklist['Genre']=='Fiction']
fic
```

```python
# fic_year = fic[['Genre', 'Year']].groupby('Year').count()
# fic_year

fic_year = fic.groupby('Year').count()['Genre']
```

```python
fic_year.plot(kind = 'bar', figsize= (14,5))
plt.ylabel('No of Fiction Books Published') # add to x-label to the plot
plt.xlabel('Year of Book Published') # add y-label to the plot
plt.title('Number of Fiction book published in one year') # add title to the plot

plt.show()
```

# Non Fiction

```python
n_fic = booklist[booklist['Genre']== 'Non Fiction']
n_fic.head()
```

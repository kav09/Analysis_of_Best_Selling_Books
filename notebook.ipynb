```python
t_genre = booklist.groupby('Genre').count()
t_genre
```

```python
# Fiction Vs Non fIction book
plt.title('Number of Fiction Vs Non Fiction Books', y=1.15) 

plt.axis('equal') 

# add legend
plt.legend(labels=t_genre.index, loc='upper left')


t_genre['Total'].plot(kind = 'pie',figsize =(8,5),autopct='%.0f%%',startangle=90,shadow=True)
plt.show()
```
